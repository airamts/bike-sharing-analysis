### Cleaning Data

#### Delete unused table

```python
del hour_df
```

#### Delete unused column

```python
drop_col = ["instant"]

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

#### Detailing column name

```python
day_df.rename(columns={
    "dteday": "dateday",
    "yr": "year",
    "mnth": "month",
    "weathersit": "weather_category",
    "hum": "humidity",
    "cnt": "count"
}, inplace=True)

day_df.head()
```

#### Detailing number in season, year, month day, weather category

```python
day_df["season"] = day_df["season"].map({
    1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"
})
day_df["year"] = day_df["year"].map({
    0: "2011", 1: "2012"
})
day_df["month"] = day_df["month"].map({
    1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun",
    7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"
})
day_df["weekday"] = day_df["weekday"].map({
    0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"
})
day_df["weather_category"] = day_df["weather_category"].map({
    1: "Clear/Partly Cloudy",
    2: "Misty/Cloudy",
    3: "Light Rain/Snow",
    4: "Heavy Rain/Snow"
})
```

#### Change false dtype

```python
day_df["dateday"] = pd.to_datetime(day_df.dateday)
day_df["season"] = day_df.season.astype("category")
day_df["year"] = day_df.year.astype("category")
day_df["month"] = day_df.month.astype("category")
day_df["holiday"] = day_df.holiday.astype("category")
day_df["weekday"] = day_df.weekday.astype("category")
day_df["workingday"] = day_df.workingday.astype("category")
day_df["weather_category"] = day_df.weather_category.astype("category")

day_df.head()
```

## Visualization & Explanatory Analysis

### Question 1: How is the trend of bike rentals in 2011 and 2012? Which year shows better performance?

```python
day_df["month"] = pd.Categorical(day_df["month"], categories=
    ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    ordered=True)

monthly_counts = day_df.groupby(by=["year", "month"]).agg({
    "count": "sum"
}).reset_index()

# Set the style
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
sns.lineplot(
    data = monthly_counts,
    x = "month",
    y = "count",
    hue = "year",
    palette = "Paired",
    marker = "o"
)
plt.xlabel(None)
plt.ylabel(None)
plt.title("Bike Rental Trends in 2011 and 2012")
plt.legend(title = "Year", loc = "upper right")
plt.gca().xaxis.grid(False)
plt.show()
```

### Question 2: What is the comparison of bike rentals between casual users and registered users?

```python
user_counts = day_df.groupby(by="year").agg({
    "casual": "sum",
    "registered": "sum"
}).reset_index()

# Set the style
sns.set(style="whitegrid")

# Create a figure and axis
plt.figure(figsize=(10, 5))

# Plot the data using seaborn"s barplot
sns.barplot(data = user_counts, x = "year", y="casual", color="skyblue", label="Casual Users")
sns.barplot(data = user_counts, x = "year", y="registered", color="pink", label="Registered Users", bottom=user_counts["casual"])

# Add label and title
plt.xlabel(None)
plt.ylabel(None)
plt.title("Comparison of Bike Rental between Casual and Registered Users")

# Add a legend
plt.legend(title = "User Type", loc = "upper right")

plt.grid(False)
plt.show()
```

### Question 3: What are the patterns of bike rentals between holiday, weekday, and working day?

```python
weekday_counts = day_df.groupby(by="weekday").agg({
    "count": "sum",
}).reset_index()

sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
# Berdasarkan weekday
sns.barplot(data = day_df, x = "weekday", y = "count", color = "lightblue")
plt.title("Total of Bike Rental by Weekday")
plt.xlabel(None)
plt.ylabel(None)

plt.grid(False)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,10))

# Berdasarkan workingday
ax[0].pie(day_df.groupby("workingday")["count"].sum(), 
          labels=["Non-Workingday", "Workingday"],
          autopct='%1.1f%%', 
          colors=["skyblue", "pink"])
ax[0].set_title("Percentage of Bike Rental by Workingday")
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)

# Berdasarkan holiday
ax[1].pie(day_df.groupby("holiday")["count"].sum(), 
          labels=["Non-Holiday", "Holiday"], 
          autopct='%1.1f%%', 
          colors=["skyblue", "pink"])
ax[1].set_title("Percentage of Bike Rental by Holiday")
ax[1].set_xlabel(None)
ax[1].set_ylabel(None)

plt.tight_layout()
plt.show()
```

### Question 4: What is the relationship between temperature (temp and atemp), humidity (hum), and windspeed with the total of bike rentals?

```python
plt.figure(figsize=(12, 6))

# Scatter plot untuk "temp" vs "count"
plt.subplot(2, 2, 1)
sns.scatterplot(
    x = "temp",
    y = "count",
    data = day_df,
    alpha = 0.5
)
plt.title("Temperature vs Count")

# Scatter plot untuk "atemp" vs "count"
plt.subplot(2, 2, 2)
sns.scatterplot(
    x = "atemp",
    y = "count",
    data = day_df,
    alpha = 0.5
)
plt.title("Temperature (Feels Like) vs Count")

# Scatter plot untuk "hum" vs "count"
plt.subplot(2, 2, 3)
sns.scatterplot(
    x = "humidity",
    y = "count",
    data = day_df,
    alpha = 0.5
)
plt.title("Humidity vs Count")

# Scatter plot for windspeed vs cnt
plt.subplot(2, 2, 4)
sns.scatterplot(
    x = "windspeed",
    y = "count",
    data = day_df,
    alpha = 0.5
)
plt.title("Windspeed vs Total Rental Bikes")

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
correlation = day_df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation, dtype=bool))

sns.heatmap(
    correlation,
    annot=True,
    mask=mask,
    cmap="mako",
    center=0,
    fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

### Question 5: How do the changes in season and weather conditions influence the total of bike rentals?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x="season",
    y="count",
    hue="weather_category",
    palette = "Paired",
    data = day_df)

plt.title("Bike Rentals Based on Weather Conditions and Seasons")
plt.xlabel(None)
plt.ylabel(None)
plt.show()
```

## Conclusion

- Conclusion 1: There was a difference between the peak and valley difference in the months of the two years (2011 and 2012). The peak of 2011 was in June and the valley was in January, while the peak of 2012 was in September and the valley is the same in January. And from this visualizations, we can observe that the total number of bike rented in 2012, is entirely higher than in 2011.

- Conclusion 2: In both years (2011 and 2012), registered users significantly outnumbered casual users in bike rentals. Their segment (pink) is much larger than that of casual users (blue). Besides that, From 2011 to 2012, there was an overall increase in bike rentals for both user types. The number of bikes rented by registered users almost reached 2 million in 2012.

- Conclusion 3: The total bike rentals are fairly consistent throughout the weekday, with each day having close to 5000 rentals. Besides that, total bike rented significantly higher percentage of bike rentals occur on working days and non-holidays. Specifically, 69.6% of bike rentals happen on working days, while only 30.4% occur on non-working days. Similarly, 97.6% of bike rentals are on non-holidays, with only a small fraction (2.4%) taking place during holidays.

- Conclusion 4: From the 4 visualizations shown, it can be concluded that temp and atemp have a positive correlation with the total bike rentals, so if the temperature (temp and atemp) increases, then total bike rentals also tend to increase. On the contrary, with humidity and windspeed that have a negative correlations with total bikes rentals although very minimal (not too significant), but can still decrease the number of cycle users as humidities and wind speeds increase.

- Conclusion 5: Summer and Fall have the highest bike rentals, reaching close to 6000 rentals during clear/partly cloudy days. Conversely, the number of rentals decreases notably during light rain/snow, winter, and spring, also further diminishes during misty/cloudy weather.
